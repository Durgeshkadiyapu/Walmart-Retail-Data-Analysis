```python
from statsmodels.tsa.stattools import adfuller
```

```python
result = adfuller(df1['Weekly_Sales'])
result
```

```python
p_val=result[1]
p_val*10000
```

```python
if(p_val<0.05):
  print('accept the null hypothesis:','data is stationary')
else:
  print('accept the alternate hypothesis:','data is not stationary')
```

```python
df1.shape
```

```python
train_size = int(len(df1) * 0.8)

train = df1.iloc[:train_size]
test  = df1.iloc[train_size:]
```

```python
train
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
from statsmodels.graphics.tsaplots import plot_pacf
plot_pacf(df1['Weekly_Sales'].diff().dropna(), lags=30)
```

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(df1['Weekly_Sales'].diff().dropna(), lags=30)
```

#arima model building

```python
model = ARIMA(train['Weekly_Sales'], order=(1,0,2))
results = model.fit()

print(results.summary())
```

```python
pred = results.predict(
    start=test.index[0],
    end=test.index[-1],
    typ='levels'
)
```

```python
df1['predicted']=pred
```

```python
df1
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(train['Weekly_Sales'], label='Train')
plt.plot(test['Weekly_Sales'], label='Test', color='black')
plt.plot(pred, label='Predicted', color='red')
plt.legend()
plt.title('ARIMA Prediction vs Actual')
plt.show()
```

#Building sarimax model

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(
    train['Weekly_Sales'],
    order=(1,0,2),              # (p,d,q)
    seasonal_order=(1,1,1,52)  # (P,D,Q,s)

)

results = model.fit()
print(results.summary())
```

```python
pred = results.predict(
    start=test.index[0],
    end=test.index[-1],
    typ='levels'
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(train['Weekly_Sales'], label='Train')
plt.plot(test['Weekly_Sales'], label='Test', color='black')
plt.plot(pred, label='SARIMA Prediction', color='red')
plt.legend()
plt.title('SARIMA Prediction vs Actual')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(test, pred))
print("Auto ARIMA RMSE:", rmse)
```

#building sarimax model using given columns also

```python
df2=df[df['Store']==40]
```

```python
df2
```

```python
exog_cols = [
    'Temperature',
    'Fuel_Price',
    'CPI',
    'Unemployment',
    'Holiday_Flag'
]

X = df2[exog_cols]
y = df2['Weekly_Sales']
```

```python
X
```

```python
y
```

```python
train_size = int(len(df2) * 0.8)

y_train = y.iloc[:train_size]
y_test  = y.iloc[train_size:]

X_train = X.iloc[:train_size]
X_test  = X.iloc[train_size:]
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

sarimax_model = SARIMAX(
    y_train,
    exog=X_train,
    order=(1,0,2),
    seasonal_order=(1,1,1,52),
    enforce_stationarity=False,
    enforce_invertibility=False
)

sarimax_results = sarimax_model.fit()
print(sarimax_results.summary())
```

```python
sarimax_pred = sarimax_results.predict(
    start=y_test.index[0],
    end=y_test.index[-1],
    exog=X_test,
    typ='levels'
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,6))
plt.plot(y_train, label='Train')
plt.plot(y_test, label='Test', color='black')
plt.plot(sarimax_pred, label='SARIMAX Prediction', color='red')

plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.title('SARIMAX: Train vs Test vs Prediction')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error


rmse = np.sqrt(mean_squared_error(test, sarimax_pred))
print("Auto ARIMA RMSE:", rmse)
```
